```python
file_obj=open('sentences.txt')
```

```python
data_list = file_obj.readlines()
file_obj.close()
```

```python
data_list = [line.lower() for line in data_list]
```

```python
for line in data_list: print line.strip()
```

```python
import re
data_array = [re.split('[^a-z]', line) for line in data_list]
```

```python
def rmv(l):
    l = [w for w in l if w!='']
    return l

data_array = [rmv(l) for l in data_array]
```

```python
data_array 
```

```python
#проверяет наличие значения w в словаре d
def contain(w,d):
    for i in range(len(d)):
        if d[i]==w: return True
    return False    
```

```python
words = dict()
i=0
for line in data_array:
    for word in line:
            if not(contain(word,words)):
                words[i]=word
                i+=1
```

```python
words
```

```python
n=len(data_list)
d=len(words)
```

```python
import numpy as np
```

```python
#возвращает количество вхождений слова w в предложение s
def count(w,s):
    k=0
    for i in s:
        if i==w: k+=1
    return k
```

```python
matrix = np.empty((n, d))
for i in range(n):
    for j in range(d):
        matrix[i][j]=count(words[j],data_array[i])
```

```python
matrix.shape
```

```python
from scipy import spatial
dist=np.empty((n))
for i in range(1,n):
    dist[i] = scipy.spatial.distance.cosine(matrix[0],matrix[i])
dist
```

```python
minimum=dist[1]
ind_min=1
for i in range(1,n):
    if dist[i]<minimum:
        pr_min=minimum
        pr_ind_min=ind_min
        minimum=dist[i]
        ind_min=i
```
